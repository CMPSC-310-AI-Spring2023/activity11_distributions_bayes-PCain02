## Binomial Distribution

The random binomial randomly genertates integers based on the number of trials, the size, and the probability of success.

While experimenting I noticed that as I decreased the probabilty of success the numbers got lower. When I increased the probability of success the numbers got higher. The generated numbers tended to stick around the percentage of the sucess probability.

```python
import numpy as np

n=100 # number of trials
p=0.2 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

In this case the samples are uniformly distributed based on the given low and high values. 

When I changed the high value it changed the maxiumum number a sample could be. When I experimented with the low value it changed the lowest possible value.

```python
import numpy as np

uniform = np.random.uniform(low=5, high=20,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

Another name for normal distribution is the bell curve. 

Changing the mean from 0 to 0.7 made the values get larger. And making the standard deviation larger made the difference in values more extreme.

```python
import numpy as np
mu, sigma = .7, 0.2 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

## Bayes Algorithm

Making the test_size larger made the number of total points larger.

Changing the random state to be a larger number made the number of mislabeled points to be larger.

Ref: https://scikit-learn.org/stable/modules/naive_bayes.html

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=3)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```
